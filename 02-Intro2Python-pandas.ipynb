```python
# This function is very useful to split data into a train and test set.
# These sets will be used in the training and testing of machine learning models

train_df = df.sample(frac=0.8, random_state=123) # we take 80% of the data and ensure we get the same data on each time
test_df  = df[~df.index.isin(train_df.index)]
print(df.shape)
print(train_df.shape)
print(test_df.shape)
```
