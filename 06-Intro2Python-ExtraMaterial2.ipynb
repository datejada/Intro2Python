```python
# import packages
import pandas as pd
```

```python
# read in data
df_employees = pd.read_csv('./datasets/employees.csv')
display(df_employees)
df_locations = pd.read_csv('./datasets/locations.csv')
display(df_locations)
```

```python
# concatenate dataframes
df_employees_locations = pd.concat([df_employees, df_locations], axis=1)
df_employees_locations
# not so much what we want
```

```python
# merge dataframes
# 1. inner join (default) - only rows that match in both dataframes
df_merged = pd.merge(df_employees,
                     df_locations,
                     on='company',
                    )
df_merged
```

```python
# 2. outer join - all rows from both dataframes
df_merged = pd.merge(df_employees,
                     df_locations,
                     on='company',
                     how='outer'
                    )
df_merged
```

```python
# 3. left join - all rows in left dataframe, only rows that match in right dataframe
df_merged = pd.merge(df_employees,
                     df_locations,
                     on='company',
                     how='left'
                    )
df_merged
```

```python
# 3. right join - all rows in right dataframe, only rows that match in left dataframe
df_merged = pd.merge(df_employees,
                     df_locations,
                     on='company',
                     how='right'
                    )
df_merged
```

```python
# join indicator - tells you where the rows come from
df_merged = pd.merge(df_employees,
                     df_locations,
                     on='company',
                     how='outer',
                     indicator=True
                    )
df_merged
```

```python
# merge validation - make sure you have the same number of rows after merging
df_merged = pd.merge(df_employees,
                     df_locations,
                     on='company',
                     how='inner',
                     validate='m:1' # 1:1, 1:m, m:1, m:m
                    )
df_merged
```
